```python
# Extracting Shufrat text
# Searching dimensions must be text # if they are dimensions must explode
# Searching layer is set in this section
# Searching text must havve %%C (fi) in it
text_arr = []
insertion_x = []
insertion_y = []
for text in acad.iter_objects('Text'):
    if text.Layer == "Dimensione-UTS" or  text.Layer == "dim armim":
        # print("Found layer")
        if "%%C" in text.TextString:
            
            # print("Found text")
            text_arr.append(text.TextString)
            insertion_x.append(text.InsertionPoint[0])
            insertion_y.append(text.InsertionPoint[1])

# create pandas dataframe
# d = {'col1': [1, 2], 'col2': [3, 4]}
# df = pd.DataFrame(data=d)
data = {'text': text_arr, 'pos.X': insertion_x, 'pos.Y': insertion_y}
shufrat_text = pd.DataFrame(data = data)
```

```python
shufrat_text
```

```python
# Text pattern and extract from text
# Actual text .pattern // import pandas as pd
shufrat_text_ = shufrat_text
shufrat_text_['shufra'] = [int(i.split('%%C')[1].split(',')[0]) for i in shufrat_text['text']]
shufrat_text_['gjatesia'] = [int(i.split(', L=')[1][:-2]) for i in shufrat_text['text']]
shufrat_text_['cope'] = [int(eval(i.split('%%C')[0].split(r'\P')[1])) for i in shufrat_text['text']]
```

```python
# Defining Element start string
element_start_string = ["KOLONA", "MURI"]
```

```python
# Exctracting Element list
# Element text layer = "dim element"
el_list = []
el_insertion_x = []
el_insertion_y = []
for text in acad.iter_objects('Text'):
    if text.Layer == "dim element":
        if '%%U' in text.TextString:
            try:
                # print(text.TextString[3:].split(' ')[0])
                # temp_text = text.TextString[3:].split(' ')[0]
                if text.TextString[3:].split(' ')[0] in element_start_string:
                    el_list.append(text.TextString[3:].split(' ')[0] + ' '+ text.TextString[3:].split(' ')[1])
                    el_insertion_x.append(text.InsertionPoint[0])
                    el_insertion_y.append(text.InsertionPoint[1])
            except:
                pass
```

```python
el_data = {'name': el_list, 'pos.X': el_insertion_x, 'pos.Y': el_insertion_y}
Elementet = pd.DataFrame(data = el_data)
```
