```python
# Heirarchical clustering # HC 
# Results very similar to K-means
# 2 types of hierarchical clustering
# Agglomerative & Divisive
# Divisive is the reverse of agglomerative
# can search Divisive in the reverse

# Agglomerative HC
# Step1 : Make each datapoint a single-point cluseter. that forms N Clusters
# S2: Take 2 closest datapoints and make them one cluster - N-1 Clusters
# S3: repeat step 2 - N-2
# S4: repeat step 3 until there is only 1 cluster left
# Closest clusters # may choose varies of distances Ecludian

# Euclidian distance: D=sqrt((x2-x1)**2+(y2-y1)**2)
# distance of clusters:
# Option1: Closest points distance
# o2: Furthest points distane
# o3: Average distance
# o4: Centroid distance

# HC Stores memory of dengorgrams (clusters of each step)

# Dengorgrams//How are they created
# Dengorgramps is made of X-points axes; Y is distances
# Dendorgram is the memory of closest points in each step
# a line is placed with the height of distance of points
# yAxis represents disimilarity
# Dendrogram is used to define number of clusters
# may set dissimilarity threshhold
# seperation is made between horizontal lines with the largest distances


##############################
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Importing the dataset
# must make UTS/Laptop varibales
ws_sub = r"Hierarchical Clustering"
ws = r"Data"
data_file = "Mall_Customers.csv"

data_file_path = os.path.join(os.path.join(os.path.join(ws, ws_sub), "data"),data_file)

dataset = pd.read_csv(data_file_path)
# dataset

# splitting the dataset
X = dataset.iloc[:, [3, 4]].values
```

```python
# Using the dengogram to find the optimal number of clusters
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
# method='ward' // minimise within cluster variance 
# like in clustering k-means we minimize sum of squares
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidian distances')
plt.show()
```

```python
# Fitting hierarchical clustiring to the mall dataset
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
# Visualising the data
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], color = 'red', label = 'Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], color = 'blue', label = 'Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], color = 'green', label = 'Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], color = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], color = 'magenta', label = 'Cluster 5')
#plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color = 'gray', s = 200, label = 'Centroids')
plt.title('Clusters of clients')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show()
```
