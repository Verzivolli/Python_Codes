```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Importing the dataset
# must make UTS/Laptop varibales
ws = r"Data\Random Forest Regression"
data_file = os.path.join(os.path.join(ws,"data"),'Position_Salaries.csv')
dataset = pd.read_csv(data_file)
dataset
```

```python
# splitting the dataset
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
# y
```

```python
# Fitting Random forest regressor
# Random forest is as having multiple decision tree regresion models predicting the same value
from sklearn.ensemble import RandomForestRegressor
# help(RandomForestRegressor)
regressor = RandomForestRegressor(n_estimators = 10)
# n_estimators // number of trees
regressor.fit(X, y)
```

```python
# Visualising the Training set Result
x_grid = np.arange(min(X), max(X), 0.01)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(X, y, color = "green")
# plt.scatter(X_test, Y_test, color = "black")
plt.plot(x_grid, regressor.predict(x_grid), color = "blue")
plt.title("random forest regression tree (SVR)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
regressor_100 = RandomForestRegressor(n_estimators = 100)
# more trees doesnt mean more steps // avarega will converge as we add tress
# n_estimators // number of trees
regressor_100.fit(X, y)
# Visualising the Training set Result
x_grid = np.arange(min(X), max(X), 0.01)
x_grid = x_grid.reshape(len(x_grid), 1)
plt.scatter(X, y, color = "green")
# plt.scatter(X_test, Y_test, color = "black")
plt.plot(x_grid, regressor_100.predict(x_grid), color = "blue")
plt.title("random forest regression tree (SVR)")
plt.xlabel("Position level")
plt.ylabel("Salary")
plt.show()
```

```python
y_pred = regressor_100.predict(6.5)
y_pred
```
